Removing Outliers of New Sampled  BY IQR Method

```python

# Removing Outliers of features have strong correlation

v3_fraud = dfs['V3'].loc[dfs['Class'] == 1].values
q25, q75 =np.percentile(v3_fraud, 25), np.percentile(v3_fraud, 75)
print('Quartile 25: {} | Quartile 75: {}'.format(q25, q75))
v3_iqr =q75 - q25
print('iqr: {}'.format(v3_iqr))
v3_cut_off =v3_iqr * 1.5
v3_lower, v3_upper = q25 - v3_cut_off, q75 + v3_cut_off
print('Cut Off: {}'.format(v3_cut_off))
print('V3 Lower: {}'.format(v3_lower))
print('V3 Upper: {}'.format(v3_upper))

outliers = [x for x in v3_fraud if x < v3_lower or x > v3_upper]
print('Feature V3 Outliers for Fraud Cases: {}'.format(len(outliers)))
print('V3 outliers:{}'.format(outliers))
dfs = dfs.drop(dfs[(dfs['V3'] > v3_upper) | (dfs['V3'] < v3_lower)].index)
print('----' * 44)

##################################################################################################################

v14_fraud = dfs['V14'].loc[dfs['Class'] == 1].values
q25, q75 =np.percentile(v14_fraud, 25), np.percentile(v14_fraud, 75)
print('Quartile 25: {} | Quartile 75: {}'.format(q25, q75))
v14_iqr =q75 - q25
print('iqr: {}'.format(v14_iqr))
v14_cut_off =v14_iqr * 1.5
v14_lower, v14_upper = q25 - v14_cut_off, q75 + v14_cut_off
print('Cut Off: {}'.format(v14_cut_off))
print('V14 Lower: {}'.format(v14_lower))
print('V14 Upper: {}'.format(v14_upper))

outliers = [x for x in v14_fraud if x < v14_lower or x > v14_upper]
print('Feature V14 Outliers for Fraud Cases: {}'.format(len(outliers)))
print('V14 outliers:{}'.format(outliers))
dfs = dfs.drop(dfs[(dfs['V14'] > v14_upper) | (dfs['V14'] < v14_lower)].index)
print('----' * 44)

###################################################################################################################

v12_fraud = dfs['V12'].loc[dfs['Class'] == 1].values
q25, q75 = np.percentile(v12_fraud, 25), np.percentile(v12_fraud, 75)
v12_iqr = q75 - q25
v12_cut_off = v12_iqr * 1.5
v12_lower, v12_upper = q25 - v12_cut_off, q75 + v12_cut_off
print('V12 Lower: {}'.format(v12_lower))
print('V12 Upper: {}'.format(v12_upper))
outliers = [x for x in v12_fraud if x < v12_lower or x > v12_upper]
print('V12 outliers: {}'.format(outliers))
print('Feature V12 Outliers for Fraud Cases: {}'.format(len(outliers)))
dfs = dfs.drop(dfs[(dfs['V12'] > v12_upper) | (dfs['V12'] < v12_lower)].index)
print('Number of Instances after outliers removal: {}'.format(len(dfs)))
print('----' * 44)

##################################################################################################################

v10_fraud = dfs['V10'].loc[dfs['Class'] == 1].values
q25, q75 = np.percentile(v10_fraud, 25), np.percentile(v10_fraud, 75)
v10_iqr = q75 - q25
v10_cut_off = v10_iqr * 1.5
v10_lower, v10_upper = q25 - v10_cut_off, q75 + v10_cut_off
print('V10 Lower: {}'.format(v10_lower))
print('V10 Upper: {}'.format(v10_upper))
outliers = [x for x in v10_fraud if x < v10_lower or x > v10_upper]
print('V10 outliers: {}'.format(outliers))
print('Feature V10 Outliers for Fraud Cases: {}'.format(len(outliers)))
dfs = dfs.drop(dfs[(dfs['V10'] > v10_upper) | (dfs['V10'] < v10_lower)].index)
print('Number of Instances after outliers removal: {}'.format(len(dfs)))
```

# SPLITTING THE DATASET

```python
# Splitting data into X and Y
X=dfs.drop(columns='Class')
y=dfs['Class']
```

```python
# X dataframe
X.shape
```

```python
# Y dataframe 
y.shape
```

```python
# Applying Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

# APPLYING MODELS

```python
# Importing different models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
```

```python
models={"Logistic Regression":LogisticRegression(),
        "KNeighbors":KNeighborsClassifier(),
        "SVC":SVC(),
        "Naive Bayes":GaussianNB(),
        "Decision Tree":DecisionTreeClassifier(),
        "Random Forest":RandomForestClassifier(),
        "AdaBoost": AdaBoostClassifier(),
        "XGBoost": XGBClassifier()}
```

```python
CROSS-VALIDATION SCORE OF MODELS BEFORE HYPERTUNING PARAMETERS
```

```python
# Cross validation score of models
from sklearn.model_selection import cross_val_score

for name,classifier  in models.items():
    classifier.fit(X_train,y_train)
    training_score=cross_val_score(classifier,X_train,y_train,cv=5)
    print(f"Classifier:{name} has a training score of {np.mean(training_score)*100:.2f}% accuracy")
```

# HYPERTUNING PARAMETERS WITH GRID SEARCH CV

```python
# Hypertuning of parameters of models using GridSearch CV
from sklearn.model_selection import GridSearchCV

# Logistic Regression parameters hypertuning
log_paras ={"penalty":['l1','l2'],'C':[0.001,0.01,0.1,1,10,100,1000]}
grid_search_log =GridSearchCV(LogisticRegression(),log_paras)
grid_search_log.fit(X_train,y_train)
log= grid_search_log.best_estimator_

# KNeighbors Classifier parameters hypertuning
knear_paras={"n_neighbors":list(range(2,5,1)),'algorithm':['auto','ball_tree','kd_tree']}
grid_search_knear =GridSearchCV(KNeighborsClassifier(),knear_paras)
grid_search_knear.fit(X_train,y_train)
knear= grid_search_knear.best_estimator_

# Support Vector Classifier parameters hypertuning
svc_paras ={'kernel':['rbf','poly','sigmod','linear'],'C':[0.5,0.7,0.9,1,10]}
grid_search_svc =GridSearchCV(SVC(),svc_paras)
grid_search_svc.fit(X_train,y_train)
svc= grid_search_svc.best_estimator_

# Decision tree Classifier parameters hypertuning
tree_paras ={"criterion":["gini","entropy"],"max_depth":list(range(2,4,1)),"min_samples_leaf":list(range(5,7,1))}
grid_search_tree =GridSearchCV(DecisionTreeClassifier(),tree_paras)
grid_search_tree.fit(X_train,y_train)
tree= grid_search_tree.best_estimator_

# RandomForestClassifier parameters hypertuning
rf_paras={'n_estimators':[100,200],'max_depth':[10,20]}  
grid_search_rf =GridSearchCV(RandomForestClassifier(),rf_paras)
grid_search_rf.fit(X_train,y_train)
rf= grid_search_rf.best_estimator_

# AdaBoostClassifier parameters hypertuning
abc_paras={'n_estimators':[100,200],'learning_rate':list(range(1,2,1))}  
grid_search_abc =GridSearchCV(AdaBoostClassifier(),abc_paras)
grid_search_abc.fit(X_train,y_train)
abc= grid_search_abc.best_estimator_

# XGBOOSTClassifier parameters hypertuning
xg_paras={'n_estimators':[100,200],'max_depth':[3,5]}
grid_search_xg =GridSearchCV(XGBClassifier(),xg_paras)
grid_search_xg.fit(X_train,y_train)
xg= grid_search_xg.best_estimator_

# Naives Bayes have no parameters
        
```

```python
# Hypertuning Results
print(log,'\n',knear,'\n',svc,'\n',tree,'\n',rf,'\n',abc,'\n',xg)
```

# CROSS - VALIDATION  SCORE OF MODELS AFTER HYPERTUNING

```python
# cross_val_score after hypertuning of models

# Logistic Regression
log_score=cross_val_score(log,X_train,y_train,cv=5)
print('Logistic Regression has cross validation score:',round(log_score.mean()*100,2).astype(str)+'%')


# KNeighbors Classifier
knear_score=cross_val_score(knear,X_train,y_train,cv=5)
print('KNeighbors has cross validation score:',round(knear_score.mean()*100,2).astype(str)+'%')

# Support Vector Classifier
svc_score=cross_val_score(svc,X_train,y_train,cv=5)
print('SVC has cross validation score:',round(svc_score.mean()*100,2).astype(str)+'%')

# Decision Tree Classifier
tree_score=cross_val_score(tree,X_train,y_train,cv=5)
print('Decision Tree has cross validation score:',round(tree_score.mean()*100,2).astype(str)+'%')

#Random Forest Classifier
rf_score=cross_val_score(rf,X_train,y_train,cv=5)
print('Random Forest has cross validation score:',round(rf_score.mean()*100,2).astype(str)+'%')

#AdaBoost Classifier
abc_score=cross_val_score(abc,X_train,y_train,cv=5)
print('AdaBoost has cross validation score:',round(abc_score.mean()*100,2).astype(str)+'%')


#XGBoost Classifier
xg_score=cross_val_score(xg,X_train,y_train,cv=5)
print('XGBoost has cross validation score:',round(xg_score.mean()*100,2).astype(str)+'%')

```

```python
# Cross Validation Predictions

from sklearn.metrics import roc_curve
from sklearn.model_selection import cross_val_predict

# Logistic Rgression Prediction
log_pred= log.predict(X_test)

# KNeighbors Classifier Prediction
knear_pred=knear.predict(X_test)

# Support Vector Classifier Prediction
svc_pred= svc.predict(X_test)

# Decision Tree Classifier Prediction
tree_pred=tree.predict(X_test)

# Random Forest Classifier Prediction
rf_pred=rf.predict(X_test)

#AdaBoost Classifier Prediction
abc_pred=abc.predict(X_test)

# XGBoost Classifier Prediction
xg_pred=xg.predict(X_test)
```
